# **Problem Statement**

The primary objective is to predict the output value y based on two input features x1 and x2 using a basic neural network model with the equation y=3x1+4x2. The neural network will be trained using a feed-forward and back-propagation
mechanism to minimize the error between the predicted and actual values.


# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# Activation Functions

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
def tanh(x):
  return np.tanh(x)
def relu(x):
  if x<=0:
    return 0
  else:
    return x
```

# Sample Data

```python
x1= 3
x2= 2
y_actual = 17
```

# Model Building

```python
w1= random.uniform(1,10)
w2= random.uniform(1,10)
```

```python
print(f"Initial Weights: \nw1={w1} \nw2={w2}")
```

```python
alpha = 0.01
epochs = 20
```

```python
error_history=[]
ypred_history=[]
epoch_history=[]
```

```python
print("\nTraining Progress:")
print(f"{'Epoch':<10} | {'y_pred':<10} | {'Error':<10} | {'w1':<10} | {'w2':<10}")
print("-" * 60)

for epoch in range(epochs):
  y_pred = x1*w1 + x2*w2
  sqe = (y_pred-y_actual)**2
  error_history.append(sqe)
  ypred_history.append(y_pred)
  epoch_history.append(epoch+1)

  grad_w1 = 2*(y_pred-y_actual)*x1
  grad_w2 = 2*(y_pred-y_actual)*x2

  w1=w1-(alpha*grad_w1)
  w2=w2-(alpha*grad_w2)

  print(f"{epoch+1:<10} | {y_pred:<10.4f} | {sqe:<10.4f} | {w1:<10.4f} | {w2:<10.4f}")
```

# Visualization

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epoch_history, error_history, marker='o', color='red', label='Squared Error')
plt.title('Error Progression across Epochs')
plt.xlabel('Epoch')
plt.ylabel('Squared Error')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epoch_history, ypred_history, marker='o', color='blue', label='Predicted Output (y_pred)')
plt.axhline(y=y_actual, color='green', linestyle='--', label=f'Actual Target ({y_actual})')
plt.title('Prediction Progression across Epochs')
plt.xlabel('Epoch')
plt.ylabel('Output Value')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
print(f"\nFinal Prediction: {y_pred:.4f}")
print(f"Target Value: {y_actual}")
```

After sufficient training, the predicted `y_pred` is close to the actual value
`y_actual`.

The error graph does show a declining trend, indicating that the
model is learning and improving its predictions over time
